```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# x^2 + 2x + 1
def f(x):
    return 4*x**3 - 9*x**2
```

```python
# 2x + 2
def f_prime(x):
    return 12*x**2 - 18*x
```

```python
# Visualisasi fungsi
x_dist = np.linspace(-2, 3, 500)
y = f(x_dist)
y_prime = f_prime(x_dist)

plt.plot(x_dist, y, label="f(x)")
plt.plot(x_dist, y_prime, label="f'(x)")
plt.axhline(0, color="black", linewidth=0.5)
plt.title("Contoh Fungsi dengan Local & Global Minima")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Definisikan fungsi dan turunannya
def f(x):
    return x**2 + 2*x + 1

def f_prime(x):
    return 2*x + 2

# Parameter awal
x_dist = 5.0
lr = 0.1

# Simpan history
x_history = []
y_history = []

# Iterasi gradient descent
for step in range(20):
    y = f(x_dist)
    grad = f_prime(x_dist)
    x_dist = x_dist - lr * grad

    x_history.append(x_dist)
    y_history.append(y)

# Visualisasi fungsi dan langkah gradient descent
X = np.linspace(-5, 5, 100)
Y = f(X)

plt.figure(figsize=(8,5))
plt.plot(X, Y, label="f(x) = x^2 + 2x + 1")
plt.scatter(x_history, y_history, color="red", label="Gradient Descent Steps")
plt.plot(x_history, y_history, linestyle="--", color="red")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Gradient Descent on f(x) = x^2 + 2x + 1 (NumPy Version)")
plt.legend()
plt.grid(True)
plt.show()
```
