## Bar plot
Digunakan untuk:

```python
kategori = ['Python', 'Assembly', 'Java', 'C++']
jumlah_pengguna = [100, 10, 50, 70]

plt.figure(figsize=(8, 5))
# plt.bar
plt.bar(kategori, jumlah_pengguna)

# tambah elemen
plt.title('Jumlah pengguna bahasa pemrogramman', fontsize=14, fontweight='medium')
plt.xlabel('Bahasa Pemrograman')
plt.ylabel('Jumlah pengguna')
plt.grid(axis='y', alpha=0.5, linestyle='--')

plt.show()
```

```python
kategori = ['Batch 1', 'Batch 2', 'Batch 3']
laki_laki = [40, 60, 55]
perempuan = [15, 30, 50]

x = np.arange(len(kategori)) # [0, 1, 2]
barwidth = 0.35

plt.figure(figsize=(8, 5))
# plt.bar
plt.bar(x - barwidth/2, laki_laki, width=barwidth, label='laki_laki')
plt.bar(x + barwidth/2, perempuan, width=barwidth, label='perem puan')

# tambah elemen
plt.title('Jumlah pengguna bahasa pemrogramman', fontsize=14, fontweight='medium')
plt.xlabel('Bahasa Pemrograman')
plt.ylabel('Jumlah pengguna')
plt.grid(axis='y', alpha=0.5, linestyle='--')
plt.xticks(x, kategori)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import plotly.graph_objects as go
import pandas as pd

# 1. Menggunakan data fiktif yang sama
data = {
    'Bulan': ['Jan', 'Feb', 'Mar', 'Apr', 'Mei', 'Jun', 'Jul', 'Ags', 'Sep', 'Okt', 'Nov', 'Des'],
    'Produk A': [200, 210, 220, 215, 225, 230, 235, 230, 240, 245, 250, 240],
    'Produk B': [150, 155, 160, 158, 162, 165, 170, 168, 172, 175, 180, 178],
    'Produk C': [50, 55, 60, 62, 65, 70, 150, 190, 250, 310, 380, 450], # Ada lonjakan dari Juli
    'Produk D': [100, 105, 110, 108, 112, 115, 118, 120, 122, 125, 128, 130]
}
df = pd.DataFrame(data)

# 2. Membuat visualisasi DENGAN konteks
fig_after = go.Figure()

# Warna untuk fokus dan konteks
warna_fokus = 'blue'
warna_konteks = 'lightgrey'

# Menambahkan produk lain (konteks) dengan warna abu-abu
for produk in ['Produk A', 'Produk B', 'Produk D']:
    fig_after.add_trace(go.Scatter(
        x=df['Bulan'], y=df[produk],
        mode='lines+markers', name=produk,
        line=dict(color=warna_konteks, width=2)
    ))
    # fig_before.add_trace(go.Scatter(x=df['Bulan'], y=df[produk], mode='lines+markers', name=produk))

# Menambahkan Produk C (fokus cerita) dengan warna biru dan lebih tebal
fig_after.add_trace(go.Scatter(
    x=df['Bulan'], y=df['Produk C'],
    mode='lines+markers', name='Produk C',
    line=dict(color=warna_fokus, width=4)
))

# Menambahkan anotasi untuk menunjuk ke titik cerita
fig_after.add_annotation(
    x='Jul', y=150,
    text="<b>Peluncuran ulang Produk C</b><br>memicu pertumbuhan pesat",
    showarrow=True,
    arrowhead=2,
    ax=-100,
    ay=-100
)

# Membuat judul yang bercerita dan membersihkan layout
fig_after.update_layout(
    title_text="<b>Peningkatan Penjualan Didorong oleh Produk C Sejak Peluncuran di Bulan Juli</b>",
    title_x=0.5,
    xaxis_title=None,
    yaxis_title="Jumlah Penjualan",
    template='simple_white', # Template yang lebih bersih
    showlegend=False # Legenda tidak perlu karena sudah ada fokus warna
)

fig_after.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# 1. Data fiktif yang sama
data_csat = {
    'Bulan': ['Jan', 'Feb', 'Mar', 'Apr', 'Mei', 'Jun', 'Jul', 'Ags', 'Sep', 'Okt', 'Nov', 'Des'],
    'Skor Kepuasan': [75, 76, 78, 77, 72, 70, 74, 78, 82, 85, 88, 90]
}
df_csat = pd.DataFrame(data_csat)

# 2. Membuat grafik yang BENAR-BENAR BERSIH DAN EFEKTIF
sns.set_theme(style="white")
plt.figure(figsize=(12, 7))

# Plot garis data utama
ax = sns.lineplot(
    x='Bulan',
    y='Skor Kepuasan',
    data=df_csat,
    linewidth=3,
    color='royalblue',
    zorder=5 # Pastikan garis ini di lapisan paling atas
)

# ---- INILAH PERBAIKANNYA ----

# a. Tambahkan garis referensi (baseline) di skor awal
skor_awal = df_csat['Skor Kepuasan'].iloc[0]
ax.axhline(y=skor_awal, color='grey', linestyle='--', linewidth=1.5, zorder=1)
ax.text(11.5, skor_awal, 'Skor Awal', ha='right', va='center', color='grey', fontsize=10)


# b. Judul dan Subjudul yang rapi
ax.text(x=-0.5, y=102, s='Skor Kepuasan Pulih dan Mencapai Puncak Baru',
        fontsize=18, fontweight='bold', ha='left')
ax.text(x=-0.5, y=97, s='Setelah sempat turun ke titik terendah di bulan Juni, skor berhasil melampaui level awal',
        fontsize=14, ha='left', color='dimgray')

# c. Anotasi yang strategis dengan panah
# Anotasi titik terendah
ax.annotate('Titik terendah: 70',
            xy=('Jun', 70),
            xytext=(-40, -40), # Jarak teks dari titik
            textcoords='offset points',
            arrowprops=dict(arrowstyle='->', color='dimgray'),
            ha='center', fontsize=11, color='dimgray')

# Anotasi puncak baru
ax.annotate('Puncak baru: 90',
            xy=('Des', 90),
            xytext=(0, 25), # Jarak teks dari titik
            textcoords='offset points',
            ha='center', fontsize=12, fontweight='bold', color='royalblue')

# d. Membersihkan dengan lebih bijak
sns.despine(left=True) # Hapus border atas dan kanan, tapi sisakan kiri
ax.set_xlabel('')
ax.set_ylabel('')
ax.tick_params(axis='x', labelsize=11)
ax.tick_params(axis='y', labelsize=11)

# Atur batas sumbu Y agar ada ruang "bernapas"
ax.set_ylim(65, 105)

plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Dummy data: Monthly website visits by source
data = {
    'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun'],
    'Organic': [1200, 1400, 1300, 1500, 1600, 1700],
    'Paid':    [800, 900, 950, 1000, 1050, 1100],
    'Referral':[600, 700, 650, 720, 760, 800]
}
df = pd.DataFrame(data).melt(id_vars='Month', var_name='Source', value_name='Visits')

# Visualization 1: Generic (no focus attention)
plt.figure(figsize=(8, 5))
sns.lineplot(data=df, x='Month', y='Visits', hue='Source', marker='o')
plt.title('Generic View: Equal Emphasis pada Semua Source', fontsize=14)
plt.xlabel('Month', fontsize=12)
plt.ylabel('Visits', fontsize=12)
plt.legend(title='Source')
plt.tight_layout()
plt.show()

# Visualization 2: Focus Attention (highlight Paid)
plt.figure(figsize=(8, 5))
# Plot non-Paid in gray
sns.lineplot(
    data=df[df['Source'] != 'Paid'], x='Month', y='Visits',
    hue='Source', palette=['lightgray','lightgray'], legend=False, marker='o'
)
# Highlight Paid in bold red
sns.lineplot(
    data=df[df['Source'] == 'Paid'], x='Month', y='Visits',
    color='tab:red', marker='o', linewidth=2.5, label='Paid'
)
plt.title('Focus View: Menyorot Paid Channel', fontsize=14)
plt.xlabel('Month', fontsize=12)
plt.ylabel('Visits', fontsize=12)
# Annotate peak Paid
paid_df = df[df['Source']=='Paid']
peak = paid_df.loc[paid_df['Visits'].idxmax()]
plt.annotate(
    f"Paid peak\n{peak['Visits']}",
    xy=(peak['Month'], peak['Visits']),
    xytext=('May', peak['Visits']+180),
    arrowprops=dict(arrowstyle='->', color='black')
)
plt.legend()
sns.despine(trim=True)
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# 1. Membuat data fiktif
bulan = pd.date_range('2024-01-01', periods=10, freq='MS').strftime('%B')
tiket_sebelum = [480, 510, 500, 525, 540]
tiket_sesudah = [280, 260, 255, 265, 270]
data = {
    'Bulan': bulan,
    'Jumlah Tiket': tiket_sebelum + tiket_sesudah
}
df = pd.DataFrame(data)
df['Waktu'] = pd.to_datetime(df['Bulan'] + ' 2024')

# 2. Setup Plot
sns.set_theme(style="white")
fig, ax = plt.subplots(figsize=(12, 7))

# 3. Plot Garis dan Area Utama
ax.plot(df['Waktu'], df['Jumlah Tiket'], color='#1f77b4', linewidth=2.5)
ax.fill_between(df['Waktu'], df['Jumlah Tiket'], color='#1f77b4', alpha=0.1)

# 4. Menambahkan Elemen Cerita (ini yang terpenting!)

# Garis vertikal penanda intervensi
ax.axvline(x=pd.to_datetime('2024-05-15'), color='black', linestyle='--', linewidth=1.5, ymin=0, ymax=1.0)
ax.text(pd.to_datetime('2024-05-15'), 620, 'Sistem Bantuan Otomatis\nDILUNCURKAN',
        ha='center', fontsize=11, fontweight='bold')

# Anotasi sebelum intervensi
ax.text(pd.to_datetime('2024-03-01'), 400, 'Rata-rata SEBELUM:\n~511 tiket/bulan',
        ha='center', fontsize=11, color='dimgray')

# Anotasi setelah intervensi
ax.text(pd.to_datetime('2024-08-01'), 150, 'Rata-rata SESUDAH:\n~266 tiket/bulan',
        ha='center', fontsize=11, color='#1f77b4', fontweight='bold')

# Anotasi Penurunan
penurunan = (511 - 266) / 511
ax.annotate(f'Penurunan {penurunan:.0%}',
            xy=(pd.to_datetime('2024-09-15'), 265),
            xytext=(pd.to_datetime('2024-09-15'), 400),
            arrowprops=dict(facecolor='black', arrowstyle='->', connectionstyle="arc3,rad=-0.2"),
            ha='center', fontsize=12, fontweight='bold')

# 5. Membersihkan & Memberi Judul
# Judul yang bercerita
ax.text(pd.to_datetime('2024-01-01'), 720,
        'Implementasi Sistem Bantuan Otomatis Berhasil Menekan Jumlah Tiket',
        fontsize=18, fontweight='bold', ha='left')
# Subjudul
ax.text(pd.to_datetime('2024-01-01'), 680,
        'Jumlah tiket bantuan yang diterima per bulan turun signifikan sejak peluncuran pada bulan Mei',
        fontsize=14, ha='left', color='dimgray')

# Membersihkan sumbu
sns.despine(left=True, bottom=True)
ax.set_yticks([])
ax.set_ylabel('')
ax.set_xlabel('')

# Format tanggal di sumbu X
ax.tick_params(axis='x', which='major', labelsize=11)
plt.xticks(df['Waktu'], df['Bulan'])
ax.set_ylim(0, 750)

plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Siapkan Data Cerita
data = {
    'Bulan': ['Jan', 'Feb', 'Mar', 'Apr', 'Mei', 'Jun', 'Jul'],
    'Elektronik': [120, 125, 130, 235, 225, 220, 215], # Ada lonjakan besar di April
    'Pakaian': [150, 155, 148, 152, 158, 155, 160],   # Stabil
    'Perabotan': [90, 88, 92, 95, 93, 90, 94]         # Stabil
}
# --- INI BARIS YANG DIPERBAIKI ---
# Ubah dictionary menjadi DataFrame terlebih dahulu
df_wide = pd.DataFrame(data)

# Gunakan DataFrame yang sudah dibuat untuk di-melt
df = pd.melt(df_wide, id_vars=['Bulan'], var_name='Kategori', value_name='Penjualan (Juta Rp)')
# ------------------------------------

# Mengatur urutan bulan
df['Bulan'] = pd.Categorical(df['Bulan'], categories=['Jan', 'Feb', 'Mar', 'Apr', 'Mei', 'Jun', 'Jul'], ordered=True)


# 2. Setup Plot
sns.set_theme(style="white")
fig, ax = plt.subplots(figsize=(12, 7))

# 3. Plot Garis Konteks (Kategori Lain)
sns.lineplot(data=df[df['Kategori'] != 'Elektronik'], x='Bulan', y='Penjualan (Juta Rp)',
             units='Kategori', estimator=None, color='lightgrey', linewidth=2, ax=ax)

# 4. Plot Garis Utama (Elektronik)
sns.lineplot(data=df[df['Kategori'] == 'Elektronik'], x='Bulan', y='Penjualan (Juta Rp)',
             color='deepskyblue', linewidth=3.5, marker='o', markersize=8, ax=ax)

# 5. Tambahkan Elemen Cerita dan Desain
# a. Tandai periode kampanye
ax.axvspan('Mar', 'Apr', facecolor='lightskyblue', alpha=0.2, zorder=0)
ax.text(2.5, 270, 'Kampanye Digital\nDigencarkan', ha='center', fontsize=11, fontweight='bold', color='navy')

# b. Label Langsung
ax.text(6.3, df.loc[df.index[-1], 'Penjualan (Juta Rp)'], 'Elektronik',
        ha='left', va='center', fontsize=12, fontweight='bold', color='deepskyblue')
ax.text(6.3, 160, 'Kategori Lain', ha='left', va='center', fontsize=12, color='dimgray')
ax.text(6.3, 94, '(stabil)', ha='left', va='center', fontsize=12, color='dimgray')


# c. Judul yang merupakan kesimpulan cerita
fig.text(0.05, 0.97, 'Kampanye Marketing April Memicu Lonjakan Penjualan Elektronik',
         ha='left', fontsize=18, fontweight='bold')
fig.text(0.05, 0.91, 'Penjualan kategori lain yang tidak diiklankan tetap stabil, membuktikan efektivitas kampanye.',
         ha='left', fontsize=12, color='dimgray')


# d. Declutter - Membersihkan elemen yang tidak perlu
sns.despine(left=True, bottom=True)
ax.set_xlabel('')
ax.set_ylabel('Penjualan (Juta Rupiah)', fontsize=12)
ax.grid(axis='y', linestyle='--', alpha=0.6)
ax.tick_params(axis='x', labelsize=11)
ax.set_ylim(0, 300)

plt.tight_layout(rect=[0.05, 0, 1, 0.9])
plt.show()
```

## Box Plot
Box plot (atau diagram kotak dan kumis) digunakan untuk menampilkan distribusi data numerik melalui kuartilnya, median, dan untuk mengidentifikasi potensi outlier. Ini memberikan ringkasan visual yang cepat tentang pusat, sebaran, dan simetri data, serta keberadaan nilai ekstrem.

```python
plt.figure(figsize=(8, 6))
plt.boxplot(x=df['Total Amount'], vert=False)
plt.title('Distribution of Total Amount', fontsize=16, fontname='serif')
plt.ylabel('Total Amount', fontsize=12, fontname='serif')
plt.grid(True, linestyle='--', alpha=0.7)#up
plt.show() 
```

## Subplots

```python
import matplotlib.pyplot as plt

# Data dummy selama 7 hari
hari = ['Sen', 'Sel', 'Rab', 'Kam', 'Jum', 'Sab', 'Min']
followers = [120, 130, 145, 160, 175, 190, 210]
views = [300, 400, 350, 500, 600, 550, 700]
likes = [100, 120, 110, 160, 200, 180, 250]
comments = [20, 25, 22, 35, 40, 38, 50]
engagement_rate = [0.4, 0.36, 0.38, 0.39, 0.4, 0.39, 0.43]

# Buat subplot 2x2
fig, axs = plt.subplots(2, 2, figsize=(12, 8))
fig.suptitle('Trafic Media Sosial Bagusde', fontsize=24)

# Grafik 1: Pertumbuhan followers
axs[0, 0].plot(hari, followers, marker='o')
axs[0, 0].set_title('Pertumbuhan Followers')
axs[0, 0].set_ylabel('Jumlah')
axs[0, 0].grid()


# Grafik 2: Views harian
axs[0, 1].bar(hari, views, color='skyblue')
axs[0, 1].set_title('Rata-rata Views Harian')
axs[0, 1].set_ylabel('Views')
axs[0, 1].grid()


# Grafik 3: Like vs Comment
axs[1, 0].plot(hari, likes, label='Likes', marker='o')
axs[1, 0].plot(hari, comments, label='Comments', marker='s')
axs[1, 0].set_title('Likes vs Comments')
axs[1, 0].legend()
axs[1, 0].grid()

# Grafik 4: Engagement rate
axs[1, 1].scatter(hari, engagement_rate, color='green')
axs[1, 1].set_title('Engagement Rate')
axs[1, 1].set_ylabel('Rate')
axs[1, 1].set_ylim(0, 0.5)
axs[1, 1].grid()

fig.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data dummy
hari = ['Sen', 'Sel', 'Rab', 'Kam', 'Jum', 'Sab', 'Min']

# Data minggu ini
followers_minggu_ini = [120, 130, 145, 160, 175, 190, 210]
views_minggu_ini = [300, 400, 350, 500, 600, 550, 700]
likes_minggu_ini = [100, 120, 110, 160, 200, 180, 250]
comments_minggu_ini = [20, 25, 22, 35, 40, 38, 50]
engagement_minggu_ini = [0.4, 0.36, 0.38, 0.39, 0.4, 0.39, 0.43]

# Data minggu lalu
followers_minggu_lalu = [100, 115, 130, 140, 155, 170, 185]
views_minggu_lalu = [250, 300, 320, 400, 450, 480, 550]
likes_minggu_lalu = [90, 100, 95, 130, 150, 140, 200]
comments_minggu_lalu = [15, 18, 17, 25, 30, 28, 35]
engagement_minggu_lalu = [0.35, 0.33, 0.34, 0.36, 0.37, 0.36, 0.38]

# Buat subplot
fig, axs = plt.subplots(2, 2, figsize=(12, 8))

# 1. Pertumbuhan followers
axs[0, 0].plot(hari, followers_minggu_lalu, marker='o', linestyle='--', label='Minggu Lalu')
axs[0, 0].plot(hari, followers_minggu_ini, marker='o', label='Minggu Ini')
axs[0, 0].set_title('Pertumbuhan Followers')
axs[0, 0].legend()
axs[0, 0].grid(alpha=0.3)


# 2. Views harian
x_indexes = range(len(hari))
bar_width = 0.35
axs[0, 1].bar([x - 0.175 for x in x_indexes], views_minggu_lalu, width=bar_width, label='Minggu Lalu')
axs[0, 1].bar([x + 0.175 for x in x_indexes], views_minggu_ini, width=bar_width, label='Minggu Ini')
axs[0, 1].set_xticks(x_indexes)
axs[0, 1].set_xticklabels(hari)
axs[0, 1].set_title('Rata-rata Views Harian')
axs[0, 1].grid(alpha=0.3)
axs[0, 1].legend()

# 3. Likes vs Comments
axs[1, 0].plot(hari, likes_minggu_lalu, label='Likes - Lalu', linestyle='--')
axs[1, 0].plot(hari, likes_minggu_ini, label='Likes - Ini')
axs[1, 0].plot(hari, comments_minggu_lalu, label='Comments - Lalu', linestyle='--')
axs[1, 0].plot(hari, comments_minggu_ini, label='Comments - Ini')
axs[1, 0].set_title('Likes vs Comments')
axs[1, 0].legend()
axs[1, 0].grid(alpha=0.3)

# 4. Engagement Rate
axs[1, 1].plot(hari, engagement_minggu_lalu, marker='x', linestyle='--', label='Minggu Lalu')
axs[1, 1].plot(hari, engagement_minggu_ini, marker='o', label='Minggu Ini')
axs[1, 1].set_ylim(0, 0.5)
axs[1, 1].set_title('Engagement Rate')
axs[1, 1].grid(alpha=0.3)
axs[1, 1].legend()

# Tambahkan judul utama
fig.suptitle("Perbandingan Performa Konten: Minggu Ini vs Minggu Lalu", fontsize=16, fontweight='bold')

# Atur layout agar tidak bertabrakan
plt.tight_layout(rect=[0, 0, 1, 0.97])
plt.show()
```
