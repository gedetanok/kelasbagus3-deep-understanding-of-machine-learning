## Line plot
Line chart cocok untuk:
- Melihat perubahan nilai dari waktu ke waktu
- Menampilkan tren pertumbuhan atau penurunan
- Visualisasi data berurutan atau time series

```python
# our data
x = [1, 2, 3, 4, 5]
y = [10, 20, 15, 25, 30]

# buat figure dan plot
plt.figure(figsize=(8, 4))
plt.plot(x, y)

# detail elemen visual
plt.title('Contoh Line Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')
plt.grid(True)

# tampilkan
plt.show()
```

```python
# our data
x = [1, 2, 3, 4, 5]
y2 = [15, 30, 25, 20, 20]
y = [10, 20, 15, 25, 30]

# buat figure dan plot
plt.figure(figsize=(8, 4))
plt.plot(x, y, marker='o', linestyle='-', label='MTK')
plt.plot(x, y2, marker='o', linestyle='--', label='PHYSC')

# detail elemen visual
plt.title('Contoh Line Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')
plt.grid(True)
plt.legend()

# tampilkan
plt.show()
```

## Bar chart
Diagram batang cocok untuk:
- Membandingkan nilai antar kategori
- Menampilkan frekuensi data diskrit
- Visualisasi data kategorikal

```python
kategori = ['Python', 'Java', 'C++', 'JavaScript']
jumlah_pengguna = [90, 20, 50, 70]

plt.figure(figsize=(8, 5))
plt.bar(kategori, jumlah_pengguna)

plt.title('Jumlah Pengguna Bahasa Pemrograman')
plt.xlabel('Bahasa Pemrograman')
plt.ylabel('Jumlah Pengguna (juta)')
plt.grid()

plt.show()
```

```python
import numpy as np

batch = ['Batch 1', 'Batch 2', 'Batch 3']
laki_laki = [40, 60, 55]
perempuan = [35, 50, 45]

# Membuat posisi X untuk masing-masing grup
x = np.arange(len(batch))  # [0, 1, 2]
width = 0.35  # Lebar batang

# Buat plot
plt.figure(figsize=(8, 5))
plt.bar(x - width/2, laki_laki, width=width, label='Laki-laki')
plt.bar(x + width/2, perempuan, width=width, label='Perempuan')

# Tambahan elemen visual
plt.xlabel('Batch')
plt.ylabel('Jumlah Peserta')
plt.title('Perbandingan Jumlah Peserta Laki-laki & Perempuan')
plt.xticks(x, batch)  # Label x sesuai batch
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.5)

plt.tight_layout()
plt.show()
```

## Scatter plot
Scatter plot digunakan untuk:
- Menampilkan hubungan antara dua variabel numerik
- Visualisasi pola seperti korelasi, outlier, atau kluster
- Cocok untuk analisis eksploratif (EDA)

```python
umur = [18, 21, 25, 30, 35, 40, 45, 50]
penghasilan = [1.5, 2.2, 2.8, 3.5, 4.0, 4.2, 4.5, 4.7]

plt.figure(figsize=(8, 5))
plt.scatter(umur, penghasilan)

plt.title('Hubungan umur dan penghasilan')
plt.xlabel('Umur (tahun)')
plt.ylabel('Penghasilan (juta)')
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
fisika_A = [75, 80, 85, 90, 78]
matematika_A = [70, 85, 82, 88, 80]

fisika_B = [60, 65, 70, 72, 68]
matematika_B = [62, 66, 71, 69, 65]

plt.figure(figsize=(8, 5))

plt.scatter(fisika_A, matematika_A, label='Kelas A', marker='o')
plt.scatter(fisika_B, matematika_B, label='Kelas B', marker='^')

plt.title('Perbandingan Nilai Fisika dan Matematika')
plt.xlabel('Nilai Fisika')
plt.ylabel('Nilai Matematika')
plt.grid(True)
plt.legend()
plt.tight_layout()

plt.show()
```

```python
# 1
bulan = ['Jan', 'Feb', 'Mar', 'Apr', 'Mei', 'Jun']
kelas_A = [50, 80, 65, 100, 130, 150]
kelas_B = [40, 70, 60, 90, 120, 140]

plt.figure(figsize=(8, 4))
plt.plot(bulan, kelas_A, color='blue', marker='o', linestyle='--', label='Kelas A')
plt.plot(bulan, kelas_B, color='red', marker='s', linestyle='-', label='Kelas B')

plt.title('Perkembangan Jumlah Peserta per Kelas')
plt.xlabel('Bulan')
plt.ylabel('Jumlah Peserta')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.savefig('../Week 2/challenge_1_line_comparison.png')
plt.show()
```

```python
bahasa = ['Python', 'Java', 'C++', 'JavaScript']
tahun_2024 = [85, 70, 60, 75]
tahun_2025 = [90, 72, 62, 80]

x = np.arange(len(bahasa))
width = 0.35

plt.figure(figsize=(8, 4))
plt.bar(x - width/2, tahun_2024, width, label='2024', color='skyblue')
plt.bar(x + width/2, tahun_2025, width, label='2025', color='orange')
plt.xticks(x, bahasa)
plt.title('Jumlah Pengguna Bahasa Pemrograman')
plt.xlabel('Bahasa')
plt.ylabel('Jumlah (juta)')
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.savefig('../Week 2/challenge_2_bar_comparison.png')
plt.show()
```
